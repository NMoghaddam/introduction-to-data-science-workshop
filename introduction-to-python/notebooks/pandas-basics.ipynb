# Looping over datasets

```python
import glob
```

```python
glob.glob("../data/gapminder_gdp_*.csv")
```

```python
filenames = glob.glob("../data/gapminder_gdp_*.csv")

for filename in sorted(filenames):
    data = pd.read_csv(filename, index_col='country')
    print(filename, data.min())
```
